```python
from functional import seq
import json
import re
from parsing import parse_ingredient
```

```python
recipes = seq.open("../data/recipes-raw.jsonl", encoding="utf-8").map(json.loads)
recipes.len()
```

```python
ingredients = recipes.map(lambda x: x['ingredients']).flatten().list()

ingredients[0]
```

```python
success_f = open('parse-success.txt', "w")
fail_f = open('parse-fail.txt', "w")

for x in ingredients[:300]:
    try:
        parsed = parse_ingredient(x)
        success_f.write(x + " | " + json.dumps(parsed) + "\n")
    except Exception as e:
        print(e)
        fail_f.write(x + "\n")
        
success_f.close()
fail_f.close()
```
