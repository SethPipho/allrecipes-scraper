```python
from functional import seq
import json
import re
```

```python
recipes = seq.open("recipes-raw.jsonl", encoding="utf-8").map(json.loads)
```

```python
titles = recipes.map(lambda x: x['title'])
```

```python
ingredients = recipes.map(lambda x: x['ingredients']).flatten()
ingredients.take(1)
```

```python
unit_names = ["teaspoon", "tablespoon", "pound", "ounce", "pinch", "cup", "pint", "quart", "drop", "gallon"]
unit_regex = re.compile("(" + "|".join(unit_names) + ")[es]*")
```

```python
units = (ingredients
            .map(lambda x: x.lower())
            .map(lambda x: re.sub("[0-9/]","",x))
            .map(lambda x: re.sub("[\(\)]","",x)) 
            .map(lambda x: unit_regex.sub('',x))
            .map(lambda x: x.strip())
            #.map(lambda x: unit_regex.split(x)[-1].strip())
             
        )
```

```python
units.distinct().sorted().for_each(print)
```
